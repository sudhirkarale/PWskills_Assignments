**Q1. for loop vs. while loop**

**for loop:**

- Used when you know the exact number of iterations beforehand.
- Iterates a fixed number of times based on a counter variable.

**Example:**

```python
for i in range(5):  # Loop iterates 5 times
    print(i)
```

**while loop:**

- Used when you don't know the exact number of iterations in advance.
- Continues looping as long as a certain condition is true.

**Example:**

```python
count = 0
while count < 5:
    print(count)
    count += 1  # Increment counter for termination
```

**Q2. Sum and product of first 10 natural numbers**

**Using for loop:**

```python
sum = 0
product = 1
for i in range(1, 11):  # Start from 1 for natural numbers
    sum += i
    product *= i

print("Sum:", sum)
print("Product:", product)
```

**Using while loop:**

```python
i = 1
sum = 0
product = 1
while i <= 10:
    sum += i
    product *= i
    i += 1

print("Sum:", sum)
print("Product:", product)
```

**Q3. Electricity bill calculation**

```python
def calculate_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 4.5
    elif units <= 200:
        bill = 450 + (units - 100) * 6
    elif units <= 300:
        bill = 1050 + (units - 200) * 10
    else:
        bill = 1350 + (units - 300) * 20

    return bill

units = int(input("Enter units consumed: "))
bill = calculate_bill(units)

print("Electricity bill:", bill)
```

**Q4. Cube of divisible by 4 or 5**

```python
numbers = list(range(1, 101))  # Create list 1 to 100

divisible_cubes = []
for num in numbers:
    cube = num**3
    if cube % 4 == 0 or cube % 5 == 0:
        divisible_cubes.append(num)

while loop (alternative):
    i = 1
    divisible_cubes = []
    while i <= 100:
        cube = i**3
        if cube % 4 == 0 or cube % 5 == 0:
            divisible_cubes.append(i)
        i += 1

print("Numbers with divisible cubes:", divisible_cubes)
```

**Q5. Count vowels**

```python
string = "I want to become a data scientist"
vowels = "aeiouAEIOU"
vowel_count = 0

for char in string:
    if char in vowels:
        vowel_count += 1

print("Number of vowels:", vowel_count)
```
