
**Q1:**

**Code to print data types:**

```python
print(type(list_))  # Output: <class 'list'>
print(type(array_list))  # Output: <class 'numpy.ndarray'>
```

**Explanation:**

- `list_` is a Python list, while `array_list` is a NumPy array. They have different data types due to their different structures and capabilities.

**Q2:**

**Code to print data types of elements:**

```python
for element in list_:
    print(type(element))  # Output: <class 'str'> for each element

for element in array_list:
    print(type(element))  # Output: <class 'numpy.str_'> for each element
```

**Explanation:**

- Both variables currently contain elements of type `str` (strings), even though `array_list` is a NumPy array.

**Q3:**

**Code with `dtype=int` and printing data types:**

```python
array_list = np.array(object=list_, dtype=int)

for element in list_:
    print(type(element))  # Output: <class 'str'> for each element

for element in array_list:
    print(type(element))  # Output: <class 'int'> for each element
```

**Explanation:**

- Specifying `dtype=int` converts the elements of `array_list` to integers, resulting in different data types between `list_` and `array_list`.
